```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots look cleaner
sns.set()
```

```python
# Load Dataset
df = pd.read_csv("train.csv")

# Display first 5 rows
df.head()
```

```python
# Check all values
print(df.isna())
```

```python
print(df.isna().sum())
```

```python
# checking duplicate rows
df.duplicated()[df.duplicated()==True]
```

```python
# finding number of rows and columns
print("Shape (rows, columns): ", df.shape,"\n")
print("number of rows: ", df.shape[0])
print("number of columns: ", df.shape[1])
```

```python
# viewing the data types of columns
df.dtypes
```

```python
#Changing Data type
df['SaleType'] = df['SaleType'].astype('string')
df.dtypes
```

```python
# Statistical summary
df.describe(include='all')
```

```python
#Distribution of SalesPrice
plt.figure(figsize=(8,5))
sns.histplot(df['SalePrice'], bins=20)
plt.title("Distribution of SalesPrice:")
plt.show()
```

```python
#Distribution of Overall Quality
plt.figure(figsize=(8,5))
sns.histplot(df['OverallQual'], bins=20)
plt.title("Distribution of Overall Quality:")
plt.show()
```

```python
#Average Saleprice By overall Quality
overallqual_saleprice = df.groupby('OverallQual')['SalePrice'].mean()

plt.figure(figsize=(8,5))
overallqual_saleprice.plot(kind='bar')

plt.title("Average Sale Price by Overall Quality")
plt.xlabel("Overall Quality")
plt.ylabel("Average Sale Price")

plt.show()
overallqual_saleprice
```

```python
#Average Saleprice By condition
SaleCondition_saleprice = df.groupby('SaleCondition')['SalePrice'].mean()

plt.figure(figsize=(8,5))
SaleCondition_saleprice.plot(kind='bar')

plt.title("Average Sale Price by Sale Condition")
plt.xlabel("Sale Condition")
plt.ylabel("Average Sale Price")

plt.xticks(rotation=45)
plt.show()

SaleCondition_saleprice
```

```python
#Average Saleprice by YearBuilt
yearbuilt_saleprice = df.groupby('YearBuilt')['SalePrice'].mean()

plt.figure(figsize=(10,5))
yearbuilt_saleprice.plot()

plt.title("Average Sale Price by Year Built")
plt.xlabel("Year Built")
plt.ylabel("Average Sale Price")

plt.show()

yearbuilt_saleprice
```

```python
#Sales Price Vs Garage Area Relationship
plt.figure(figsize=(8,5))
sns.scatterplot(x='SalePrice', y='GarageArea', data=df)

plt.title("Sale Price vs Garage Area")
plt.xlabel("Sale Price")
plt.ylabel("Garage Area")

plt.show()

```

```python
#Correlaiton Matrix
plt.figure(figsize=(6,4))
sns.heatmap(df[['GrLivArea', 'SalePrice']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()

```

```python
#Time Based Analysis
df['SaleDate'] = pd.to_datetime(df['YrSold'].astype(str) + '-' + df['MoSold'].astype(str))
df['Month'] = df['SaleDate'].dt.to_period('M')
monthly_saleprice = df.groupby('Month')['SalePrice'].sum()
plt.figure(figsize=(10,5))
monthly_saleprice.plot()

plt.title("Monthly Total Sale Price Trend")
plt.ylabel("Total Sale Price")

plt.show()
```
